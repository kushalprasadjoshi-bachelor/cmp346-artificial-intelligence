## Control Statements

- Break 
- Continue

````python
for i in range(10):
    if i ==6:
        break
    print(i)
````

````python
for i in range(10):
    if i ==6:
        continue
    print(i,end = "")
````

````python
for i in range(1,31):
    if i%3 != 0:
        continue
    print(i)
````

````python
a = "Hello World I am Python"
for i in a:
    if i == " ":
        break
    print(i)
````

````python
a = "Hello World I am Python"
for i in a:
    if i == " ":
        continue
    print(i,end = '')
````

## While loop
```python
while <condition>:
    <operation>
```

````python
a = 0
while a<5:
    print(a,"Hello world")
    a = a+1
````

````python
a = 5
while a>0:
    print(a,"Hello world")
    a = a-1
````

````python
i = 1
n = int(input("Enter n = "))
while i<=10:
    print(n,"*",i,"=",n*i)
    i = i+1
````

````python
s = ""
i = 0
n = int(input("Enter n = "))
while i<n:
    name = input("Enter name = ")
    phone = input("Enter phone = ")
    s = s + name + " " + phone +"\n"
    i = i+1
print(s)
````

````python
# 5! = 1*2*3*4*5
s =1
i = 1
n = int(input("Enter n = "))
while i<=n:
    s = s*i
    i = i+1
print(s)
````

````python
# break
i = 0
while i<10:
    if i ==6:
        break
    print(i,end = "")
    i = i+1
````

````python
# continue
i = 0
while i<10:
    if i ==6:
        i = i+1
        continue
    print(i,end = "")
    i = i+1
````

````python
# prime = 2,3,5,7,11,13,17,19,...
n = int(input("Enter n = "))
if n >=2:
    for i in range(2,n):
        if n%i==0:
            print(n,"is Composite no")
            break
    else:
        print(n,"is Prime no")

else:
    print("Please enter valid no.")
````
